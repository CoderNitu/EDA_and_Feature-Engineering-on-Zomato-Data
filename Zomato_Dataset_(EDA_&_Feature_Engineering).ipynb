***EDA (Exploratory Data Analysis):***

```python
# seeing the value counts of country columns to get the idea about data distribution

final_df.value_counts('Country')
```

```python
country_names = final_df.value_counts('Country').index
```

```python
country_values = final_df.value_counts('Country').values
```

```python
## Pie chart -- Top 3 countries that uses zomato

plt.figure(figsize=(17,7))
plt.pie(country_values[:3],labels=country_names[:3],autopct='%1.2f%%',radius=0.8)
plt.title('Top 3 countries that uses zomato')
plt.legend(country_names[:3],title='Countries',loc='upper right')  #  bbox_to_anchor=(1, 0.5) positions the legend to the right of the chart. The coordinates are given as (x, y), where x=1 moves it fully to the right, and y=0.5 centers it vertically relative to the plot. we can only use this when loc = 'center left'
plt.show()
```

*The seaborn library doesn’t have a built-in function for pie charts, as it's primarily used for statistical plots and data visualization that focus on distributions and relationships. However, we can still use matplotlib to create a pie chart and style it to look consistent with Seaborn’s themes.*

One more thing, that here it is not useful to use bar plot for visualization as data is not evenly distributed, it's more constracted towards top two or three contries

```python
## Bar Plot

plt.figure(figsize=(18,6))
plt.bar(country_names,country_values)

for index, value in enumerate(country_values):
  plt.text(index,value+1,f'{value}',ha='center')

plt.xlabel('countries')
plt.ylabel('values')
plt.title('Top 3 countries that uses zomato')
plt.show()
```

**Observations/Inference :**

*Zomatos' maximum orders,records or transactions are from India and after that from United States and United Kingdom.*

```python
final_df.columns
```

```python
# size() is similar to count() which is an aggregate function

Ratings = final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating count'})
```

```python
Ratings
```

**Observations/Inference :**

1. When ratings are between 4.5 to 4.9 (including)----> Excellent
2. When ratings are between 4.0 to 4.4 (including)----> Very Good
3. When ratings are between 3.5 to 3.9 (including)----> Good
4. When ratings are between 2.5 to 3.4 (including)----> Average
5. When ratings are between 1.8 to 2.4 (including)----> Poor

```python
Ratings.head()
```

```python
# It is prefer mainly to use for all the seaborn plots we will use in the notebook, here even if we not use palette parameter, it is not a problem, as for every individual visualiztions we have palette parameter in it

sns.set_theme(style='darkgrid')
```

```python
# We can use different colors for different bars by changing the color palette for this plot by using palette='dark','deep','bright',etc. We can also have shades of one color and add legend for better interpretation by using palette = 'Blues','Greens','coolwarm' and set one axis as hue

plt.figure(figsize=(12,6))
sns.barplot(x='Aggregate rating', y='Rating count', data=Ratings, palette='dark',legend=False)
plt.title('Aggregate rating vs Rating Count')
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x='Aggregate rating', y='Rating count',hue='Rating color',data=Ratings,palette=['White','Red','Orange','Yellow','Green', 'Green'])
plt.title('Aggregate rating vs Rating Count')
plt.show()
```

**Observation/Inference:**

1. No count ratings are very high
2. Maximum number of ratings are from 2.4 to 3.4

```python
# Count Plot -- (to see counts of unique data points in a particular categorical column). It just like value_count() in terms of vizualizations
# The count that we can see in the plot is not rating count but it the frequency of how often the data/colors in the rating color colums exist.

sns.countplot(x='Rating color',data=Ratings,palette=['White','Red','Orange','Yellow','Green', 'Green'])
```
